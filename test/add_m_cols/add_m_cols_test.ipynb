```python
import numpy as np
from add_m_cols import add_m_cols


# Display function for Sv_new to match MATLAB structure
def display_sv_new(Sv_new):
    for idx, sv in enumerate(Sv_new, start=1):
        print(f"Column {idx}:")
        if isinstance(sv, np.ndarray):
            if sv.size > 0:
                print(sv)
            else:
                print('[]')
        else:
            print(sv)

def test_case_1():
    print("Python Test Case 1: Mixed Observed and Unobserved Columns")
    
    # Inputs
    S = np.array([[1, 2],
                  [3, 4]])  # Shape (2, 2)
    Sv = [np.eye(2), np.eye(2)]  # List of 2 covariance matrices
    Ic = [1, 3]  # Observed columns at positions 1 and 3
    n2x = 4  # Total columns
    Isv = [1, 2]  # Covariance indices for observed columns
    
    # Function call
    S_new, Sv_new, Isv_new = add_m_cols(S, Sv, Ic, n2x, Isv)
    
    # Outputs
    print("S_new:")
    print(S_new)
    
    print("\nSv_new:")
    display_sv_new(Sv_new)
    
    print("\nIsv_new:")
    print(Isv_new)

# Run Test Case 1
test_case_1()
```
