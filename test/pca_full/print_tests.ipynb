```python
import matplotlib.pyplot as plt
import numpy as np


import matplotlib.pyplot as plt
import numpy as np

# Setup plot for testing
dsph = {
    'display': True,
    'rms': None,  # Will initialize later
    'prms': None  # Will initialize later
}

# Define the display_progress function to initialize and update plot data
def display_progress(dsph, lc):
    if dsph['display']:
        plt.clf()  # Clear the figure to prevent overplotting
        
        # Plot the 'rms' data
        dsph['rms'], = plt.plot(range(len(lc['rms'])), lc['rms'], 'r-', label='rms')
        
        # Plot the 'prms' data
        dsph['prms'], = plt.plot(range(len(lc['prms'])), lc['prms'], 'g-', label='prms')
        
        plt.xlabel('Step')
        plt.ylabel('Value')
        plt.legend()
        plt.draw()
        plt.pause(0.5)  # Pause to allow the plot to update

# Enable interactive mode
plt.ion()

# First test case with initial data
lc = {
    'rms': [0.5, 0.4, 0.35],
    'prms': [0.6, 0.5, 0.45]
}
display_progress(dsph, lc)

# Update values and display again
lc['rms'] = [0.5, 0.4, 0.35, 0.3]
lc['prms'] = [0.6, 0.5, 0.45, 0.4]
display_progress(dsph, lc)

# Turn off interactive mode and show final plot
plt.ioff()
plt.show()


```
