```python
import numpy as np
import scipy.io
import os


# Load MATLAB data and run the test using the rotate_to_pca function
def test_with_matlab_data():
    data_folder = 'data_rotateToPCA_test'
    os.makedirs(data_folder, exist_ok=True)
    data_file = os.path.join(data_folder, 'test_data.mat')

    # Load the test data from the .mat file in the specified folder
    data = scipy.io.loadmat(data_file)
    # Extract variables
    A = data['A']
    S = data['S']
    Av = [data['Av'][0, i] for i in range(data['Av'].shape[1])]
    Sv = [data['Sv'][0, i] for i in range(data['Sv'].shape[1])]
    Isv = data['Isv'][0] if 'Isv' in data and data['Isv'].size > 0 else []
    obscombj = [np.array([j[0]]) for j in data['obscombj'][0]]
    update_bias = bool(data['update_bias'][0, 0])

    # Run RotateToPCA
    dMu, A_rot, Av_rot, S_rot, Sv_rot = rotate_to_pca(A, Av, S, Sv, Isv, obscombj, update_bias)

    # Display results
    print('--- Test Using MATLAB Data ---')
    print('dMu:')
    print(dMu)
    print('A_rot:')
    print(A_rot)
    print('S_rot:')
    print(S_rot)
    print('Av_rot:')
    for i, Av_i in enumerate(Av_rot):
        print(f'Av_rot[{i}]:')
        print(Av_i)
    print('Sv_rot:')
    for j, Sv_j in enumerate(Sv_rot):
        print(f'Sv_rot[{j}]:')
        print(Sv_j)

# Run the test
test_with_matlab_data()
```

```python
import numpy as np
import scipy.io
import os


# Define the path to the data file in the specified folder
data_folder = 'data_rotateToPCA_test'
mat_filename = os.path.join(data_folder, 'test_data_LargeMatrices.mat')

def test_rotate_to_pca_large_matrices():
    # Load the test data from the .mat file
    data = scipy.io.loadmat(mat_filename)
    # Define the matrix dimensions and number of components
    n1 = 100
    n2 = 80
    ncomp = 10

    # Extract variables
    A = data['A']
    S = data['S']
    Av = [data['Av'][0, i] for i in range(data['Av'].shape[1])]
    Sv = [data['Sv'][0, i] for i in range(data['Sv'].shape[1])]
    Isv = data['Isv'][0] if data['Isv'].size > 0 else []
    obscombj = [np.array([j[0]]) for j in data['obscombj'][0]]
    update_bias = bool(data['update_bias'][0, 0])

    # Expected results from MATLAB
    expected_dMu = data['dMu']
    expected_A_rot = data['A_rot']
    expected_Av_rot = [data['Av_rot'][0, i] for i in range(data['Av_rot'].shape[1])]
    expected_S_rot = data['S_rot']
    expected_Sv_rot = [data['Sv_rot'][0, i] for i in range(data['Sv_rot'].shape[1])]

    # Run the rotate_to_pca function with the loaded data
    dMu, A_rot, Av_rot, S_rot, Sv_rot = rotate_to_pca(A, Av, S, Sv, Isv, obscombj, update_bias)

    # Display summary of results for comparison
    print("--- Test Large Matrices ---")
    
    # dMu summary
    print("dMu Summary - Python: Min =", np.min(dMu), "Max =", np.max(dMu))
    print("dMu Summary - MATLAB: Min =", np.min(expected_dMu), "Max =", np.max(expected_dMu))
    print("dMu match:", np.allclose(dMu, expected_dMu, atol=1e-4))

    # Check dimensions of A_rot and S_rot
    print("\nA_rot dimensions match:", A_rot.shape == expected_A_rot.shape)
    print("S_rot dimensions match:", S_rot.shape == expected_S_rot.shape)

    # Av_rot trace summaries
    print("\nAv_rot Summary:")
    for i, (actual, expected) in enumerate(zip(Av_rot, expected_Av_rot)):
        print(f"  Av_rot[{i}] Trace - Python:", np.trace(actual), "MATLAB:", np.trace(expected))

    # Sv_rot trace summaries
    print("\nSv_rot Summary:")
    for j, (actual, expected) in enumerate(zip(Sv_rot, expected_Sv_rot)):
        print(f"  Sv_rot[{j}] Trace - Python:", np.trace(actual), "MATLAB:", np.trace(expected))

    # Basic checks on dimensions
    if A_rot.shape == (n1, ncomp) and S_rot.shape == (ncomp, n2):
        print("\nSuccess: Rotated matrices have correct dimensions.")
    else:
        print("\nFailure: Rotated matrices have incorrect dimensions.")

    # Check for NaNs
    if not np.isnan(A_rot).any() and not np.isnan(S_rot).any():
        print("Success: No NaN values in rotated matrices.")
    else:
        print("Failure: NaN values found in rotated matrices.")

# Run the test
test_rotate_to_pca_large_matrices()
```

```python
import numpy as np

def test_rotate_to_pca_invalid_input_types():
    print("--- Test Invalid Input Types ---")

    # Initialize valid parameters for testing
    n1, n2, ncomp = 5, 4, 2
    np.random.seed(0)  # For reproducibility
    A = np.random.randn(n1, ncomp)
    S = np.random.randn(ncomp, n2)
    Av = [np.eye(ncomp) for _ in range(n1)]
    Sv = [np.eye(ncomp) for _ in range(n2)]
    Isv = []
    obscombj = [np.array([i]) for i in range(1, n2 + 1)]
    update_bias = True

    # Define a list of invalid inputs with descriptions
    invalid_inputs = [
        ("Invalid A", "invalid_A", Av, S, Sv, Isv, obscombj, update_bias, True),
        ("Invalid Av", A, "invalid_Av", S, Sv, Isv, obscombj, update_bias, True),
        ("Invalid S", A, Av, "invalid_S", Sv, Isv, obscombj, update_bias, True),
        ("Invalid Sv", A, Av, S, "invalid_Sv", Isv, obscombj, update_bias, True),
        ("Invalid Isv", A, Av, S, Sv, "invalid_Isv", obscombj, update_bias, False),
        ("Invalid obscombj", A, Av, S, Sv, Isv, "invalid_obscombj", update_bias, False),
        ("Invalid update_bias", A, Av, S, Sv, Isv, obscombj, "invalid_update_bias", False)
    ]

    # Test each invalid input case
    for description, test_A, test_Av, test_S, test_Sv, test_Isv, test_obscombj, test_update_bias, expect_error in invalid_inputs:
        print(f"Testing {description}...")
        try:
            # Attempt to run with invalid input
            dMu, A_rot, Av_rot, S_rot, Sv_rot = rotate_to_pca(
                test_A, test_Av, test_S, test_Sv, test_Isv, test_obscombj, test_update_bias
            )
            if expect_error:
                print(f"No Error: {description} did not throw an error as expected.\n")
            else:
                print(f"Success: {description} did not throw an error, matching MATLAB's behavior.\n")
        except Exception as e:
            if expect_error:
                # Expected error message
                print(f"Expected Error for {description}: {e}\n")
            else:
                print(f"Unexpected Error for {description}: {e}\n")

# Run the test
test_rotate_to_pca_invalid_input_types()
```
