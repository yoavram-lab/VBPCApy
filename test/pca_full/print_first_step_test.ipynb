```python
# Python Test Case 1: verbose=True with valid prms

import numpy as np
def print_first_step(verbose, rms, prms=np.nan):
    """
    Prints the initial RMS errors based on verbosity settings.

    Parameters:
    - verbose (bool): Flag to determine whether to print the information.
    - rms (float): RMS error.
    - prms (float, optional): Probe RMS error (default is NaN).
    """
    if not verbose:
        return

    # Start constructing the output string with the RMS error
    output = f"Step 0: rms = {rms:.6f}"

    # If prms is a valid number (not NaN), append it to the output string
    if not np.isnan(prms):
        output += f" ({prms:.6f})"

    # Print the constructed output string
    print(output)

# Test Case 1 Execution
print("Python Test Case 1: verbose=True with valid prms")
print_first_step(verbose=True, rms=0.123456, prms=0.654321)
print("Expected Output: Step 0: rms = 0.123456 (0.654321)\n")
```
