```python
import numpy as np
import scipy.io
import os
from pca_full import rotate_to_pca

# Define the path to the data folder
data_folder = 'data_rotateToPCA_test'

# List of test files to check
test_files = [
    'test_data_NcompOne_9683.mat',
    'test_data_EmptyIsv_8129.mat',
    'test_data_NonEmptyIsv.mat',
    'test_data_NoBiasUpdate.mat',
    'test_data_LargeMatrices.mat'
]

# Function to calculate deviation ranges for S_rot and Sv_rot
def calculate_deviation_ranges():
    s_rot_deviation_ranges = []
    sv_rot_deviation_ranges = []
    
    for test_file in test_files:
        # Load the MATLAB results from the .mat file
        mat_filename = os.path.join(data_folder, test_file)
        data = scipy.io.loadmat(mat_filename)

        # Extract variables from MATLAB data
        A = data['A']
        S = data['S']
        Av = [data['Av'][0, i] for i in range(data['Av'].shape[1])]
        Sv = [data['Sv'][0, i] for i in range(data['Sv'].shape[1])]
        Isv = data['Isv'][0] if data['Isv'].size > 0 else []
        obscombj = [np.array(data['obscombj'][0, i].flatten()) for i in range(data['obscombj'].shape[1])]
        update_bias = bool(data['update_bias'][0, 0])

        # Expected results from MATLAB
        expected_S_rot = data['S_rot']
        expected_Sv_rot = [data['Sv_rot'][0, i] for i in range(data['Sv_rot'].shape[1])]

        # Run the rotate_to_pca function with the loaded data in Python
        _, _, _, S_rot, Sv_rot = rotate_to_pca(A, Av, S, Sv, Isv, obscombj, update_bias)

        # Calculate deviations for S_rot
        s_rot_deviation = np.abs(S_rot - expected_S_rot)
        s_rot_min_deviation = np.min(s_rot_deviation)
        s_rot_max_deviation = np.max(s_rot_deviation)
        s_rot_deviation_ranges.append((s_rot_min_deviation, s_rot_max_deviation))

        # Calculate deviations for each Sv_rot matrix
        sv_rot_deviation = [np.abs(sv_python - sv_matlab) for sv_python, sv_matlab in zip(Sv_rot, expected_Sv_rot)]
        sv_rot_min_deviation = min(np.min(dev) for dev in sv_rot_deviation)
        sv_rot_max_deviation = max(np.max(dev) for dev in sv_rot_deviation)
        sv_rot_deviation_ranges.append((sv_rot_min_deviation, sv_rot_max_deviation))

    # Display the deviation ranges for each test case
    for i, test_file in enumerate(test_files):
        print(f"--- Deviation Ranges for {test_file} ---")
        print(f"S_rot Deviation Range: Min = {s_rot_deviation_ranges[i][0]:.4f}, Max = {s_rot_deviation_ranges[i][1]:.4f}")
        print(f"Sv_rot Deviation Range: Min = {sv_rot_deviation_ranges[i][0]:.4f}, Max = {sv_rot_deviation_ranges[i][1]:.4f}")
        print()

    # Calculate the median of the deviations
    s_rot_median_min = np.median([dev[0] for dev in s_rot_deviation_ranges])
    s_rot_median_max = np.median([dev[1] for dev in s_rot_deviation_ranges])
    sv_rot_median_min = np.median([dev[0] for dev in sv_rot_deviation_ranges])
    sv_rot_median_max = np.median([dev[1] for dev in sv_rot_deviation_ranges])

    # Display the median deviation ranges
    print(f"--- Median Deviation Ranges ---")
    print(f"S_rot Median Deviation Range: Min = {s_rot_median_min:.4f}, Max = {s_rot_median_max:.4f}")
    print(f"Sv_rot Median Deviation Range: Min = {sv_rot_median_min:.4f}, Max = {sv_rot_median_max:.4f}")

# Run the deviation calculation
calculate_deviation_ranges()
```
