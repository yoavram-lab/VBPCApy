```python
# Test Case 1

import numpy as np
from miscomb import miscomb
M = np.array([
    [1, np.nan, 1, np.nan],
    [2, np.nan, 2, np.nan],
    [3, np.nan, 3, np.nan]
])

verbose = True

nobscomb_p, obscombj_p, Isv_p = miscomb(M, verbose)

# Display results
print('Python Results:')
print(f'nobscomb_p: {nobscomb_p}')
print('obscombj_p:')
print(obscombj_p)
print('Isv_p:')
print(Isv_p)
```
