```python
# compute_rms_test_case2.py

# Test 2: Sparse X
print('Test 2: Sparse X')

import numpy as np
import scipy.io
import scipy.sparse as sp
from compute_rms import compute_rms  # Ensure compute_rms and errpca_pt are properly implemented

# Load data from 'test_data_case2.mat' generated by MATLAB
data = scipy.io.loadmat('test_data_compute_rms/test_data_case2.mat')

# Reconstruct the sparse matrix X
X_sparse = data['X_sparse']
iX = X_sparse['i'][0, 0].flatten() - 1  # Adjust for MATLAB's 1-based indexing
jX = X_sparse['j'][0, 0].flatten() - 1
vX = X_sparse['v'][0, 0].flatten()
X_shape = X_sparse['size'][0, 0].flatten().astype(int)
X = sp.csr_matrix((vX, (iX, jX)), shape=(X_shape[0], X_shape[1]))

# Load other matrices
A = data['A']
S = data['S']
M = data['M']
ndata = int(data['ndata'][0, 0])
rms_matlab = data['rms_matlab'][0, 0]

# Ensure data types are appropriate
A = np.array(A, dtype=np.float64)
S = np.array(S, dtype=np.float64)
M = np.array(M, dtype=np.float64)

# Compute RMS using Python function
rms_python, errMx_python = compute_rms(X, A, S, M, ndata)

print(f'RMS (Python): {rms_python}')
print(f'RMS (MATLAB): {rms_matlab}')

# Compare RMS values
tolerance = 1e-10
if abs(rms_python - rms_matlab) < tolerance:
    print('Test passed: RMS values match.')
else:
    print('Test failed: RMS values do not match.')
```
